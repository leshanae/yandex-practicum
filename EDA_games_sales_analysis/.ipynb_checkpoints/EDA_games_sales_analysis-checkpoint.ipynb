# Задание и описание данных
Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.<br>
Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.


Описание данных
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.<br>
Данные за 2016 год могут быть неполными.

# 3. Анализ

## 3.1. Сколько игр выпускалось в разные годы. Важны ли данные за все периоды?

```python
games.pivot_table(index='year_of_release', values='name', aggfunc='count')\
    .sort_values(by='year_of_release')\
    .plot(kind='bar', figsize=(16, 6), xlabel='Год', ylabel='Выпущено игр, шт.', legend=False)\
    .set_title('Выпуск игр по годам')
plt.show()
```

На графике мы видим, что до начала девяностых выпускалось не так много игр, до ста штук в год. Но с начала девяностых и до конца 2000-х наблюдался рост с небольшими искажениями в 1998, 2002 и 2003 годах.<br>
В связи с изменениями на рынке игростроя (например, появление долгоиграющих f2p игр и большее распространение MMO) количество выпускаемых игр резко снизилось, вдвое между пиковым значением и началом периода, который имеет смысл рассматривать в нашем исследовании - 2014-2016 гг.

## 3.2 Как менялись продажи по платформам? За какой характерный срок появляются новые и исчезают старые платформы?
Для начала посмотрим рейтинг по выручке.

```python
games.pivot_table(index='platform', values='total_sales', aggfunc='count').sort_values(by='total_sales')\
    .plot(kind='bar', figsize=(16, 6), xlabel='Платформа', ylabel='Суммарная выручка, млн долл', legend=False)\
    .set_title('Суммарная выручка платформ за весь период')
plt.show()
```

```python
# Создаём список потенциально прибыльных платформ
top_5_sales_list = games.groupby('platform')['total_sales'].sum().sort_values(ascending=False).head().index.tolist()

# Передаём список в срез
top_5_sales = games.query('platform in @top_5_sales_list').pivot_table(index='platform', columns='year_of_release', values='total_sales', aggfunc='sum')
```

```python
plt.figure(figsize=(20,8)).set_facecolor('lightgrey')
sb.heatmap(data=top_5_sales, annot=True, fmt='.1f')\
    .set(xlabel="Год выпуска", ylabel="Платформа", title='Выручка по годам для топ-5 самых прибыльных платформ')
plt.show()
```

Как видим, продолжительность жизни платформы составляет 10-12 лет.

## 3.7-3.8 Как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков?
Построим графики и посчитаем корреляции между доходом и рейтингами для всех платформ. Мы можем рассмотреть только числовые рейтинги, поэтому рейтинг ESRB не рассматриваем.

```python
for platform_name in games_act['platform'].unique():
    fig_scatter, (scatter_1, scatter_2) = plt.subplots(2, 1, figsize=(20,7))
    fig_scatter.set_facecolor('lightgrey')
    plt.suptitle(platform_name, y=1.05, fontsize=18)

    scatter_1.scatter(y='critic_score', x='total_sales', data = games_act[games_act['platform'] == platform_name])
    scatter_1.set_title('Рейтинг критиков')
    scatter_2.scatter(y='user_score', x='total_sales', data = games_act[games_act['platform'] == platform_name])
    scatter_2.set_title('Рейтинг пользователей')

    plt.show()

    print(f'Корреляция между доходом платформы и рейтингом критиков для {platform_name}:',
          round(games_act.query('platform == @platform_name and critic_score.notna()')['critic_score']\
                .corr(games_act.query('platform == @platform_name and critic_score.notna()')['total_sales']), 2))
    print(f'Корреляция между доходом платформы и рейтингом пользователей для {platform_name}:',
          round(games_act.query('platform == @platform_name and user_score.notna()')['user_score']\
                .corr(games_act.query('platform == @platform_name and user_score.notna()')['total_sales']), 2))
    print('')
```

Три общих наблюдения, которые можно рассмотреть, например, на платформе X360:
1. Почти для всех платформ работает закон: среди хитов нет игр с рейтингом ниже "средне-высокого".<br>
2. В рамках усов коробки слабая корреляция всё же прослеживается (что и подтверждает подсчёт корреляции продаж от рейтингов), которая потом весьма резко перерастает в предыдущее утверждение.<br>
3. В целом, пользовательские рейтинг ниже рейтингов критиков. Однако игры с низким уровнем продаж (нишевые) имеют своих преданных фанатов, оценивающих любимую игру выше, чем критики. В принципе, это можно увидеть на опрактически отсутствующей (а где-то и отрицательной) корреляции пользовательского рейтинга и дохода - увеличение этого рейтинга совсем не означает роста продаж.<br>

Отдельно стоит отметить DS, PSP и Wii, по которым данных за этот период, можно сказать, нет вообще, и выводы не сделать.<br>
Интересной смотрится PS Vita: люди покупают игры с абсолютно разным рейтингом почти одинаково часто. Есть даже игра с откровенно плохим рейтингом, но при этом в рамках платформы её продажи сичтаются хитовыми и - более того - максимальными для платформы. Парадокс!<br>

Однозначно могу сказать, что в магазин имеет смысл закупать игры для PS4, XOne, как и было выявлено ранее. Всё остальное требует рассмотрения. Не стоит сбрасывать со счетов Wii и устаревающую PS3, а также найдутся коллекционеры для коробок от PC.<br>
3DS - консоль весьма немолодая, вышла в 2011 и, в соответствии с моим исследованием, ещё имеет запас в 5 лет. Однако она не так популярна в России, поэтому если и закупать игры под неё, то в очень небольшом количестве и только самые популярные для этой платформы игры.

## 3.9 Общее распределение игр по жанрам

```python
# Таблицы для счётчика и продаж по жанрам
genre_count = games_act.pivot_table(index='genre', values='name', aggfunc='count').sort_values(by='name')
genre_sales = games_act.pivot_table(index='genre', values='total_sales', aggfunc='sum').sort_values(by='total_sales')
```

Делаю вывод, что лучше всего закупать игры четырёх самых попудярных жанров: экшен, шутер, РПГ, спортивные. При небольшом количестве закупки они могут принести наибольший доход.

# 5. Проверка гипотез

Перед проверкой гипотез я обнаружил, что в столбце пользовательского рейтинга были пропуски, они будут мешать. Я решили их не заполнять, поэтому при проверке гипотез исключу пустые значения из выборок.<br>
На языке человека, смотрящего на рейтинги, "одинаковы ли" или "примерно равны" может означать, например, разницу в одну десятую при десятибальной шкале: рейтинг 6,5 и рейтинг 6,4 это примерно одинаковые оценки (разница в 1%), как правило люди говорят, что рейтинги одинаковые даже при разнице в 2-3%. Поэтому большая точность мне не требуется и я не буду предъявлять высоких требований по критическому уровню значимости, поэтому для p-value возьму стандартные 5%.<br>

## 5.1 Одинаковы ли средние пользовательские рейтинги платформ Xbox One и PC?

Платформы не зависят друг от друга, поэтому совокупности также независимы. У нас нет оснований полагать, что дисперсии равны.<br>

$H_0$: Средние пользовательские рейтинги user_score платформы XBox = cредним пользовательским рейтингам (user_score) платформы PC<br>
$H_a$: Средние пользовательские рейтинги user_score платформы XBox ≠ cредним пользовательским рейтингам (user_score) платформы PC<br>
alpha = 0.05<br>

Нулевая гипотеза может быть перефразирована так: если вероятность ошибиться в отвержении нулевой гипотезы ниже критического значения, нулевая гипотеза отвергается.
Применительно к нашему вопросу: если мы не отвергаем нулевую гипотезу (она соответствует изначальному вопросу), то ответ "да, одинаковы".

```python
results = st.ttest_ind(games_act.query('platform == "XOne"')['user_score'],
                       games_act.query('platform == "PC"')['user_score'],
                       equal_var = False, nan_policy='omit')

alpha = 0.05

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу') # рейтинги НЕодинаковы
else:
    print('Не получилось отвергнуть нулевую гипотезу') # рейтинги одинаковы
```

Средние значения рейтингов по платформам Xbox One и PC - одинаковы, т.к. p-value превысил alpha.

## 5.2 Разные ли средние пользовательские рейтинги жанров Action и Sports?

Платформы не зависят друг от друга, поэтому совокупности также независимы. У нас нет оснований полагать, что дисперсии равны.

За нулевую гипотезу всегда брать именно равенство, то гипотезы таковы:<br>
$H_0$: Средние пользовательские рейтинги (user_score) жанра Action = cредним пользовательским рейтингам (user_score) жанра Sports<br>
$H_a$: Средние пользовательские рейтинги (user_score) жанра Action ≠ cредним пользовательским рейтингам (user_score) жанра Sports<br>
alpha = 0.05<br>

Нулевая гипотеза может быть перефразирована так: если вероятность ошибиться в отвержении нулевой гипотезы ниже критического значения, нулевая гипотеза отвергается.
Применительно к нашему вопросу: если мы не отвергаем нулевую гипотезу (она соответствует изначальному вопросу), то ответ "да, рейтинги разные".

```python
results = st.ttest_ind(games_act.query('genre == "Action"')['user_score'],
                       games_act.query('genre == "Sports"')['user_score'],
                       equal_var = False, nan_policy='omit')

alpha = 0.05

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу') # рейтинги НЕодинаковы
else:
    print('Не получилось отвергнуть нулевую гипотезу') # рейтинги одинаковы
```

Средние пользовательские рейтинги жанров Action и Sports - разные, т.к. значение p-value незначительно мало в сравнении с значением alpha.
